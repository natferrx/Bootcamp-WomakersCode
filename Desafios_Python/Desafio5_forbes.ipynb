## **Desafio 5: Forbes (Natalia e Cedma)**

**Definição do problema:** Você está trabalhando em um jornal e o
editor quer que você responda as seguintes questões:
1. Qual a empresa mais valiosa da lista da Forbes 2000?
2. Construa uma tabela com as TOP 10 empresas mais lucrativas da
lista.
3. Qual a média de valores de empresa das cinco categorias mais
valiosas?

**Código do dataset:** Forbes2000

**Dicas:**
- Método .max()
- Método nlargest()
- Método .groupby()

```python
# instalar a biblioteca para importação dos dados
!pip install pydataset
from pydataset import data
```

```python
# Carregar o dataset Forbes2000
forbes2000 = data('Forbes2000')
```

```python
# Visualizar o conjunto de dados
forbes2000.head()
```

```python
# Qual a empresa mais valiosa da lista da Forbes 2000?
empresa_mais_valiosa = forbes2000.loc[forbes2000['marketvalue'].idxmax()]

print("Empresa mais valiosa:")
print(empresa_mais_valiosa[['name', 'marketvalue']])
```

```python
# Construir uma tabela com as TOP 10 empresas mais lucrativas da lista
top_10_lucrativas = forbes2000.nlargest(10, 'profits')[['name', 'profits']]
print("\nTOP 10 empresas mais lucrativas:")
print(top_10_lucrativas)
```

```python
# Qual a média de valores de empresa das cinco categorias mais valiosas?
top_categorias_valiosas = forbes2000.groupby('category')['marketvalue'].mean().nlargest(5)
print("\nMédia de valores de empresa das cinco categorias mais valiosas:")
print(top_categorias_valiosas)
```
