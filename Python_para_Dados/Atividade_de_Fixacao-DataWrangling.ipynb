```python
import pandas as pd
import numpy as np

df = pd.read_csv("collected research people.csv")
```

```python
df.head()
```

```python
# deletar a coluna que nao tem utilidade
df = df.drop(columns=['Unnamed: 0'])

# Arrumar o nome das Colunas
novas_cols = {'Endereerço': 'Endereço',
              'phone': 'Telefone',
              'anual_income': 'Renda Anual',
              'AAge': 'Idade',
              'n_filhos': 'N Filhos',
              'Food_restrictions': 'Restrição Alimentar',
              'diariamente_sun blocker': 'Usa protetor solar diariamente',
              'streaming': 'Streaming'
              }

df.rename(columns=novas_cols,inplace=True)

df.head()
```

```python
# remover caracteres especiais da coluna nome

df.Nome = df.Nome.str.replace('[^a-zA-Zà-úÀ-Ú0-9 ]','', regex=True)
df.head()
```

```python
#corrigir coluna de telefone

df.Telefone = df.Telefone.str.replace('[^0-9() +-]','*', regex=True)
df.loc[df.Telefone.str.contains('[*]', na=True), 'Telefone'] = np.nan
df.Telefone.str.contains('[*]', na=True)
df.Telefone
```

```python
df['Estado Civil'] = df['Estado Civil'].replace({'S': 'Solteiro',
                                 "C": "Casado",
                                 "D": "Divorciado",
                                 "V": "Viúvo"})

df
```

**Cenário 2:** 
Uma empresa chamada Hillo, é uma empresa que está estudando o mercado e quer encontrar uma parceria com uma empresa de Streaming (Netflix, Disney Plus etc), mas gostaria de saber quais as empresas dariam o maior retorno de investimento. Eles podem fazer análise de machine learning também com esses dados.


```python
df.columns
```

```python
df_hillo = df.copy()
df_hillo = df_hillo.drop(columns=['Moradia', 
                                  'Restrição Alimentar',
                                  'Usa protetor solar diariamente',
                                  'Vai na Praia Mensalmente', 
                                  'Faz Academia',
                                  'Compra e Lê Livros Todos os Anos', 
                                  'Quantidade de Livros Comprados',
                                  'Autor Favorito'])
df_hillo.head()
```

```python
# A coluna Streaming tem um "e", pois pode ser mais de uma plataforma que a pessoa usa. Por isso vamos usar um separador.

df_hillo['Streaming'].str.get_dummies(sep=' e ')
```

```python
df_hillo.Streaming = df_hillo.Streaming.replace({"Todos": "Globo Play e Netflix e HBO e Disney Plus e Crunchyroll"})

df_hillo.Streaming = df_hillo.Streaming.replace({"Netflix e Disney": "Netflix e Disney Plus"})
```

```python
df_hillo_strem = df_hillo['Streaming'].str.get_dummies(sep=' e ')
df_hillo = pd.concat([df_hillo, df_hillo_strem], axis=1)
df_hillo.head()
```
