**Agregação de dados:**

Utilizando o arquivo: concurso.csv



```python
df_concurso = pd.read_csv('concurso.csv')

df_concurso.head()
```

7. Qual a distribuição de escolaridade por estado?



```python
df_concurso[['Estado','Escolaridade','Número de Inscrição']].groupby(['Estado','Escolaridade']).count()
```

8. Qual é a porcentagem de pessoas com deficiência?



```python
df_concurso['Deficiência'].value_counts(normalize=True) * 100
```

9. Qual o número de inscrições por estado?

```python
df_concurso[['Número de Inscrição','Estado']].groupby('Estado').count()
```

10. Quantas pessoas são do estado do Piauí?

```python
df_concurso[df_concurso['Estado'] == 'PI']['Número de Inscrição'].count()
```

12. Quantas pessoas nasceram antes de 1995?

```python
df_concurso['Data de Nascimento'] = pd.to_datetime(df_concurso['Data de Nascimento'])
df_antes_95 = df_concurso.loc[df_concurso['Data de Nascimento'] < np.datetime64('1995')]
df_antes_95.head(10)
```

13. Quais foram os primeiros 10 inscritos?

```python
df_10primeiras = df_concurso.set_index('Data de Inscrição').sort_index()[:10]
df_10primeiras
```
