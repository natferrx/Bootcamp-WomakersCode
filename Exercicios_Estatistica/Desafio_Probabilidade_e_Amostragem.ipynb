**QUESTAO 1 -** Considere pessoas fluentes em inglês, qual a probabilidade complementar? Ou seja, qual a probabilidade de escolhermos uma pessoa aleatória e ela não ser fluente em inglês. Considere fluente quem tem o nível avançado.

**Raciocício e comentários prévios** \
probabilidade complementar = 1 - probabilidade de algo

- amostra: conjunto total de pessoas do arquivo populacao_brasileira.csv
- calcular a probabilidade de algo
- quantas pessoas tem na nossa amostra? Resposta: TP
- quantas pessoas falam ingles no nivel avancado? Resposta: FE
- probabilidade de ser fluente em ingles PFE = FE/TP
- complementar: CO = 1 - PFE

```python
# Número total de pessoas na amostra
TP = len(df)

# Número de pessoas que falam inglês no nível avançado
FE = len(df[df['nível de proficiência em inglês'] == 'Avançado'])

# Probabilidade de ser fluente em inglês (PFE)
PFE = FE / TP

# Probabilidade complementar (CO)
CO = 1 - PFE

print("Probabilidade de não ser fluente em inglês (probabilidade complementar):", CO)
```

**QUESTAO 2** - Se uma pessoa escolhida aleatoriamente for de Alagoas ou do Pará, qual é a probabilidade de ela ter uma renda superior a 5 mil reais?

**Raciocício e comentários prévios** \
- qual a amostra que analise: pessoas que moram nos estados do Para e de Alagoas
- quantas pessoas tem na nossa amostra? Resposta: TP_Para_Alagoas
- dentro da nossa amostra filtrar as pessoas que tem renda igual ou maior que 5 mil reais. Resposta: TP_Para_Alagoas_5_ou_mais_mil_reais_renda

**QUESTAO 3** - Descubra a probabilidade de uma pessoa ter ensino superior completo no estado do Amazonas. Qual a probabilidade da quinta pessoa que você conversar, que é amazonense, ter ensino superior completo?

**Raciocício e comentários prévios** \
- amostra em questao: restringindo a nossa analise para apenas pessoas que estejam no estado do Amazonas. Resposta: P_Amazonas
- preciso saber quantas pessoas com ensino superior completo na nossa amostra (no estado do Amazonas) possuem nivel superior completo e pos graduacao. Resposta: P_Amazonas_Ensino_Superior_Completo_e_Pos_Graduacao

- primeira probabilidade
PEFETIVA =  P_Amazonas_Ensino_Superior_Completo_e_Pos_Graduacao / P_Amazonas

- PCOMPLEMENTAR = 1 - P_Amazonas_Ensino_Superior_Completo_e_Pos_Graduacao / P_Amazonas

- P1, P2, P3, P4, P5*

PCOMPLEMENTAR * PCOMPLEMENTAR * PCOMPLEMENTAR * PCOMPLEMENTAR * PEFETIVA

```python
# Filtrando a amostra para incluir apenas pessoas do estado do Amazonas
amostra_amazonas = df[df['estado'] == 'AM']

# Número total de pessoas na amostra no estado do Amazonas
P_Amazonas = len(amostra_amazonas)

# Filtrando a amostra para incluir apenas pessoas com ensino superior completo e pós-graduação
amostra_amazonas_ensino_superior_pos_graduacao = amostra_amazonas[(amostra_amazonas['escolaridade'] == 'Superior') & (amostra_amazonas['nível de proficiência em inglês'] == 'Avançado')]

# Número de pessoas com ensino superior completo e pós-graduação na amostra do estado do Amazonas
P_Amazonas_Ensino_Superior_Completo_e_Pos_Graduacao = len(amostra_amazonas_ensino_superior_pos_graduacao)

# Probabilidade efetiva (PEFETIVA)
PEFETIVA = P_Amazonas_Ensino_Superior_Completo_e_Pos_Graduacao / P_Amazonas

# Probabilidade complementar (PCOMPLEMENTAR)
PCOMPLEMENTAR = 1 - PEFETIVA

# Probabilidade de a quinta pessoa ser a primeira com ensino superior completo
P1_P2_P3_P4_P5 = PCOMPLEMENTAR ** 4 * PEFETIVA

print("Probabilidade da quinta pessoa ser a primeira com ensino superior completo:", P1_P2_P3_P4_P5)
```

**QUESTAO 4** - Considerando a renda das pessoas do nosso conjunto, podemos dizer que a renda de uma pessoa brasileira está na sua maioria em que faixa (faça faixa de 1.500 reais)? Qual é a sua função densidade de probabilidade?

**Raciocício e comentários prévios** \
- bins = 1500
- dividir as faixas salariais de 1500 e 1500
renda a: [0 - 1500.00]
renda b: ]1500.01 - 3000]
renda c: ]3000 - 4500]
renda d: ]4500 - 6000]
- contar o numero de pessoas em cada faixa

```python
import numpy as np

# Definindo os intervalos de faixas salariais
bins = np.arange(0, df['renda'].max() + 1500, 1500)

# Criando os rótulos para as faixas salariais
labels = ['Renda A', 'Renda B', 'Renda C', 'Renda D']

# Dividindo as rendas em faixas usando pd.cut e contando o número de pessoas em cada faixa
renda_por_faixa = pd.cut(df['renda'], bins=bins, labels=labels, include_lowest=True).value_counts().sort_index()

# Determinando a faixa em que a maioria das pessoas se encontra
faixa_majoritaria = renda_por_faixa.idxmax()

# Calculando a função densidade de probabilidade
funcao_densidade_probabilidade = renda_por_faixa / len(df)

print("Número de pessoas em cada faixa salarial:\n", renda_por_faixa)
print("\nFaixa salarial em que a maioria das pessoas se encontra:", faixa_majoritaria)
print("\nFunção densidade de probabilidade:\n", funcao_densidade_probabilidade)

import matplotlib.pyplot as plt

# Plotando o histograma das faixas salariais
plt.figure(figsize=(10, 6))
plt.hist(df['renda'], bins=bins, edgecolor='black', alpha=0.7)
plt.title('Histograma das Faixas Salariais')
plt.xlabel('Faixa Salarial (em reais)')
plt.ylabel('Número de Pessoas')
plt.grid(True)
plt.xticks(bins, rotation=45)
plt.tight_layout()
plt.show()

# Plotando a função densidade de probabilidade
plt.figure(figsize=(10, 6))
plt.bar(renda_por_faixa.index, funcao_densidade_probabilidade, color='skyblue', alpha=0.7)
plt.title('Função Densidade de Probabilidade das Faixas Salariais')
plt.xlabel('Faixa Salarial (em reais)')
plt.ylabel('Probabilidade')
plt.grid(axis='y')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

**QUESTAO 5** - Calcule a média e a variância da renda da amostra. Depois faça a distribuição normal, inclua o gráfico.

**Raciocício e comentários prévios** \
RENDA_MEDIA = SOMA(TODAS AS RENDAS)/NUMERO DE PESSOAS
VARIANCIA = SOMA((RENDA DE CADA INDIVIDUO - RENDA_MEDIA)^2)

```python
from scipy.stats import norm

# Calculando a média da renda da amostra
renda_media = df['renda'].mean()

# Calculando a variância da renda da amostra
renda_variancia = df['renda'].var()

# Criando uma distribuição normal com base na média e na variância
dist_normal = norm(renda_media, renda_variancia ** 0.5)  # O desvio padrão é a raiz quadrada da variância

# Gerando valores para o eixo x
x = np.linspace(renda_media - 3 * (renda_variancia ** 0.5), renda_media + 3 * (renda_variancia ** 0.5), 1000)

# Calculando os valores de densidade de probabilidade para os valores de x
y = dist_normal.pdf(x)

# Plotando o gráfico da distribuição normal
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribuição Normal', color='red')
plt.title('Distribuição Normal da Renda da Amostra')
plt.xlabel('Renda (em reais)')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print("Média da renda da amostra:", renda_media)
print("Variância da renda da amostra:", renda_variancia)
```

**QUESTAO 6** - Primeiro considere a probabilidade encontrada no nosso conjunto de pessoas com escolaridade de pós-graduação. Considerando a amostra de população brasileira com 1 milhão de habitantes, qual a probabilidade de encontrarmos 243 mil pessoas com pós-graduação?

**Raciocício e comentários prévios** \
ETAPA 1:
- amostra sera o conjunto total da populacao, dentro do arquivo populacao_brasileira.csv PPT
- filtrar por pessoas dentro da nossa amostra que tenham pos graduacao, PPT_POS_GRADUCAO
- prababilidade de selecionar uma pessoa com pos graduacao e:
PROBABILIDADE_POS_GRAD = PPT_POS_GRADUCAO / PPT


ETAPA 2:
- novo tamanho da amostra (aumentando a mesma) = n = 1 milhao

- sucesso = 243 mil pessoas com pos graduacao

- usando a probabilidade que foi calculada para a amostra na primeira etapa = PROBABILIDADE_POS_GRAD

```python
# Total de pessoas na população brasileira (amostra)
PPT = len(df)

# Total de pessoas com pós-graduação na população
PPT_POS_GRADUACAO = len(df[df['escolaridade'] == 'Pós-graduação'])

# Probabilidade de selecionar uma pessoa com pós-graduação na amostra
PROBABILIDADE_POS_GRAD = PPT_POS_GRADUACAO / PPT

# Novo tamanho da amostra (1 milhão de habitantes)
n = 1000000

# Sucesso (243 mil pessoas com pós-graduação)
sucesso = 243000

# Usando a probabilidade calculada para determinar a probabilidade de encontrar 243 mil pessoas com pós-graduação em uma amostra de 1 milhão de habitantes
from scipy.stats import binom

probabilidade_243mil = binom.pmf(sucesso, n, PROBABILIDADE_POS_GRAD)

print("Probabilidade de encontrar 243 mil pessoas com pós-graduação em uma amostra de 1 milhão de habitantes:", probabilidade_243mil)
```

**QUESTAO 7** - Somando as densidades nós temos a função de densidade acumulada. Considerando a coluna ‘Escolaridade’ faça a função de densidade acumulada discreta para cada nível de escolaridade.

**Raciocício e comentários prévios** \
valor = 0

fundamental = d_fun

valor final = 1 quando se tem o valor acumulado (100%)

**QUESTAO 8** - Qual a margem de erro amostral da proporção populacional considerando a proporção de pessoas com nível de inglês intermediário?

**Raciocício e comentários prévios** \
- primeiro analisar quantas pessas tem nivel de ingles intermediario

- p complementar (1 - prob pessoas com nivel de ingles intermediario)

- confianca: 0.95

- calcular o nosso z dada a confianca definida

margem de erro >>>> colar a formula

`Z * np.sqrt((p * (1 - p)) / n)`

```python
import scipy.stats as stats

# Número de pessoas com nível de inglês intermediário
pessoas_intermediario = len(df[df['nível de proficiência em inglês'] == 'Intermediário'])

# Proporção de pessoas com nível de inglês intermediário
p_intermediario = pessoas_intermediario / len(df)

# Proporção complementar
p_complementar = 1 - p_intermediario

# Nível de confiança
confianca = 0.95

# Calculando o valor crítico Z para o nível de confiança especificado
Z = stats.norm.ppf((1 + confianca) / 2)

# Tamanho da população
n = len(df)

# Calculando a margem de erro amostral
margem_erro = Z * np.sqrt((p_intermediario * p_complementar) / n)

print("Margem de erro amostral da proporção populacional considerando o nível de inglês intermediário:", margem_erro)
```

**QUESTAO 9** - Calcula a renda da população. Qual a probabilidade de encontrar 60
pessoas com uma renda mil reais superior à média?


**Raciocício e comentários prévios** \
- amostra = 60 pessoas

- primeiro calcular a media da renda da populacao (populacao_brasiliera.csv)

- 1 pessoa : probabilidade de ter renda >= 1000 da media (media + 1000)

- 60 pessoas: probabilidade para 60 pessoas dentro to nosso total de pessoas da amostra (populacao_brasiliera.csv) baseada na probabilidade que calculamos para 1 pessoa (calculo acima)

```python
from scipy.stats import binom

# Calcular a média da renda da população
media_renda = df['renda'].mean()

# Determinar a probabilidade de uma pessoa ter uma renda pelo menos mil reais superior à média
probabilidade_individual = 1 - stats.norm.cdf(media_renda + 1000, media_renda, df['renda'].std())

# Número de pessoas na amostra
n_amostra = 60

# Calcular a probabilidade de 60 pessoas, dentro de uma amostra de 60, terem uma renda pelo menos mil reais superior à média
probabilidade_60_pessoas = binom.pmf(60, n_amostra, probabilidade_individual)

print("Probabilidade de encontrar 60 pessoas com uma renda mil reais superior à média:", probabilidade_60_pessoas)
```

**QUESTAO 10** - Qual a probabilidade de escolhermos alguém do Sudeste que seja
homem, com ensino fundamental e com renda maior que 2 mil reais por mês?

**Raciocício e comentários prévios** \
- ser do sudeste: RJ, SP, MG, ES >>> amostra populacao_brasileira.csv
- sexo = masculino >>> amostra populacao_brasileira.csv
- escolaridade = ensino fundamental >>> amostra populacao_brasileira.csv
- renda > 2 mil reais >>> amostra populacao_brasileira.csv

P_FINAL = P_SUDESTE * P_HOMEM * P_ENSINO_FUNDAMENTAL * P_RENDA_MAIOR_2_MIL

Probabilidade final sera a combinacao das probabilidades individuais de cada occorencia vir a ser satisfeita

```python
# Calcular a probabilidade de selecionarmos alguém do Sudeste
prob_sudeste = len(df[df['estado'].isin(['RJ', 'SP', 'MG', 'ES'])]) / len(df)

# Calcular a probabilidade de selecionarmos alguém do sexo masculino
prob_homem = len(df[df['sexo'] == 'M']) / len(df)

# Calcular a probabilidade de selecionarmos alguém com ensino fundamental
prob_ensino_fundamental = len(df[df['escolaridade'] == 'Fundamental']) / len(df)

# Calcular a probabilidade de selecionarmos alguém com renda maior que 2 mil reais por mês
prob_renda_maior_2mil = len(df[df['renda'] > 2000]) / len(df)

# Calcular a probabilidade final multiplicando todas as probabilidades individuais
probabilidade_final = prob_sudeste * prob_homem * prob_ensino_fundamental * prob_renda_maior_2mil

print("Probabilidade de escolhermos alguém do Sudeste que seja homem, com ensino fundamental e com renda maior que 2 mil reais por mês:", probabilidade_final)
```
